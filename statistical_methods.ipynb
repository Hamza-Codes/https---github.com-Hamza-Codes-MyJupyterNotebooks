## Tests and their Types

1. Parametric Tests
   
      * More reliable results
      * First we have to make assumptions
 
2. Non-Parametric Tests
   
   * Less reliable results
   * Calculates the rank of data
   * No need to make assumptions


### Step - 1 : Check the normality of the data

Tests to be used: 

1. Shapiro-Wilk Test
   * Specifici (Reliable) 

2. Kolmogorov - Smirnov Test
   * General (Less reliable)

### Step - 2 : Check the Homogeneity of the Data

The variance of the variable in the data are equal

Tests to be used: 

1. Levene's Test

### Step - 3 : Purpose of your data analysis

Know the purpose of your research question

   * Comparison and Difference
      * Atleast two groups
  
   * Relationship and correlation
       * Correlation
       * Causation
       * Connection
       * Prediction


### Step - 4 : Data Type

Know the data type you are working with.

* Categorical
  * Qualitative
  * No numerical meaning
  * Represented by texts (eg. character, factors)
  * Yes or No
  * 0 or 1
  * True or False

* Continous
  * Quantitative
  * Numerical
  * Amount
  * Number
  * Age
  * Height

### Step - 5 : Statistical Tests

Choose a statistical Test from the main families

1. Chi-Squared
   * Purpose: Comparison
   * Data: Categorical only  


2. T-test/ANOVA
   * Purpose: Comparison
   * Data: Categorical and continous data  


3. Correlation
   * Purpose: Relationship
   * Data: Continous only
  
#### Types of correlation test:

1. Pearson's correlation (one-independent and one-dependent Variable)
2. Regression (one-independent and one-dependent Variable) - This tells us a specific mathematical equation that describes the relationship. Helps in predicting the missing values or future values.
