### Purposes of Data Manipulation

- consistent data

- project related data

- create more value from data

- remove or ignore data

Here are some general steps on how to approach a data maniupulation task:

1. Raw data may come from a data server or cloud or anywhere.

2. It is first cleaned and filtered to get we really need. 

3. We also ignore and delete some data. 

4. The data is tidied and cleaned up and may be combined with another dataset to make it one. Maybe one dataset is from cloud and the other is local. 

5. We can also split the data based on some argument or basis. 

6. We also deal with missing values or call the missing values. We might have some other information to calculate the missing values. 

7. Remove or merge the duplicate values.

8. We can also extract data from images which is called scrapping. We do it via Pandas.

#### Data frames in Panads

DataFrame is a 2-dimensional labeled data structure with columns of potentially different types. You can think of it like a spreadsheet or SQL table, or a dict of Series objects. It is generally the most commonly used pandas object.

```python
# Object creation

s = pd.Series([1,3,5,7,np.NaN,8,9])
s
```

```python
# Series for dates
dates = pd.date_range("20130101",periods=6)
dates
```

```python
# creating dataframe - converting the date series into data frame (rows and coloumns or like worksheet in excel)
dates = pd.date_range("
",periods=20)
dates
df = pd.DataFrame(np.random.randn(20,4), index=dates,columns=list("ABCD"))
df
```

```python
# creating dataframe - Data frame having a dictionary inside it. A dictionary has a key and a value in the form of {"key":value} 

df2 = pd.DataFrame(
    {
        "A" :1.0,
        "B": pd.Timestamp("20130101"),
        "C": pd.Series(1, index=list(range(4)),dtype="float32"), # lists the float numbers
        "D": np.array([3] * 4,dtype="int32"), # lists the integer numbers
        "E": pd.Categorical(["test","train","test","train"]), # lists all the categorical data/strings
        "F": "foo",
    }
)
df2
```
